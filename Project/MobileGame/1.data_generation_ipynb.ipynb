#가상 데이터 생성

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# 1. 유저 수
num_users = 5000
```

```python
# 2. 랜덤 데이터 생성
np.random.seed(42)
user_ids = [f'U{str(i).zfill(5)}' for i in range(num_users)]
install_dates = [datetime.today() - timedelta(days=random.randint(1, 60)) for _ in range(num_users)]
platforms = np.random.choice(['iOS', 'Android'], size=num_users)
countries = np.random.choice(['US', 'KR', 'JP', 'DE', 'FR'], size=num_users)
session_counts = np.random.poisson(lam=10, size=num_users)
total_spend = np.random.exponential(scale=20, size=num_users).round(2)
ad_sources = np.random.choice(['Facebook', 'Google', 'Organic', 'TikTok'], size=num_users)
last_session_days = np.random.randint(0, 30, size=num_users)
in_game_events = np.random.randint(0, 50, size=num_users)
retention_1d = np.random.choice([0,1], size=num_users, p=[0.3,0.7])
retention_7d = np.random.choice([0,1], size=num_users, p=[0.5,0.5])
retention_30d = np.random.choice([0,1], size=num_users, p=[0.8,0.2])
cohorts = [date.strftime('%Y-%W') for date in install_dates]
```

```python
# 3. DataFrame 생성
users_df = pd.DataFrame({
    'user_id': user_ids,
    'install_date': install_dates,
    'platform': platforms,
    'country': countries,
    'session_count': session_counts,
    'total_spend': total_spend,
    'ad_source': ad_sources,
    'last_session_days': last_session_days,
    'in_game_events': in_game_events,
    'retention_1d': retention_1d,
    'retention_7d': retention_7d,
    'retention_30d': retention_30d,
    'cohort': cohorts
})

users_df.to_csv('/content/drive/MyDrive/MobileGame/users.csv', index=False)
print("users.csv 저장 완료!")
```

```python
# 4. A/B 캠페인 데이터
ab_campaign = pd.DataFrame({
    'ad_group': ['A', 'B'],
    'impressions': [10000, 12000],
    'clicks': [500, 600],
    'installs': [200, 250],
    'revenue': [1000, 1300]
})
ab_campaign.to_csv('/content/drive/MyDrive/MobileGame/ab_campaign.csv', index=False)
print("ab_campaign.csv 저장 완료!")
```
