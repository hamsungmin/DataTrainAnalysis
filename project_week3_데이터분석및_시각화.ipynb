### 목표4
'공정명'이 '조립검사'이고 '시작시간'과 '종료시간'이 모두 2022-08-02 08:00:00 ~ 2022-08-02 09:00:00 사이인 공정의 '공정시간'의 평균을 구하여 print문으로 출력해주세요

```python
target_name = "조립검사"
min_time = datetime.strptime("2022-08-02 08:00:00", "%Y-%m-%d %H:%M:%S")
max_time = datetime.strptime("2022-08-02 09:00:00", "%Y-%m-%d %H:%M:%S")

def parse_datetime(value):
  for fmt in ["%Y-%m-%d %H:%M:%S.%f", "%Y-%m-%d %H:%M:%S"]:
    try:
      return datetime.strptime(value, fmt)
    except ValueError:
      continue
  return None

df['시작시간'] = df['시작시간'].apply(lambda x: parse_datetime(str(x)))
df['종료시간'] = df['종료시간'].apply(lambda x: parse_datetime(str(x)))

print(round(df.loc[(df["공정명"] == "조립검사") & (df["시작시간"] >= min_time) & (df["종료시간"] <= max_time), "공정시간"].mean(), 2))
```

```python
#작업자 와 설비 공정시간을 비교
#작업자와 설비 공정시간을 비교한다
#장비별 제품별 공정명별 3가지로 구분하여 분석 진행한다
group_data = df.groupby(["장비이름", "제품명", "공정명", "공정종류"])
group_data = group_data["공정시간"].mean().reset_index()
display(group_data)
#조립에 공정 시간이 작업자는 제품A, B, C 동일하게 소요하지만 설비는 제품 C > A > B순으로 시간 소요한다
#반대로 가공 공정은 설비는 동일한 공정 시간이 소요되는데 작업자는 소요시간에 차이가 있
#가공검사 공정은 설비, 작업자 공정시간이 비슷하다
```

```python
#그룹으로 합쳐진 데이터를 갖고 설비, 작업자 공정시간이 가장 큰 장비이름, 제품명, 공정명, 공정시간차이 을 표기

#설비, 작업자를 한 row에 표기되도록 작업 진행 -> 이름 바꾸기
equipment = group_data[group_data["공정종류"] == "설비"].rename(columns={"공정시간": "설비시간"})
worker = group_data[group_data["공정종류"] == "작업자"].rename(columns={"공정시간": "작업자시간"})
#이름 바꾼것을 merge 한다 한 row에 보이도록 하고 공정시간차이를 추가한다
group_data_merge = pd.merge(equipment, worker, on=["장비이름", "제품명", "공정명"], how="inner")
group_data_merge["공정시간차이"] = abs(group_data_merge["설비시간"] - group_data_merge["작업자시간"])

group_data_merge.loc[group_data_merge["공정시간차이"] == group_data_merge["공정시간차이"].max(), ["장비이름", "제품명","공정명", "공정시간차이"]]
```

```python
#가장 오래 걸리는 시간대랑 공정명을 구해라
df["시작시간"] = pd.to_datetime(df["시작시간"])
df["시간"] = df["시작시간"].dt.hour
#피벗테이블
pivot = df.pivot_table(index=["공정종류", "시간"], columns="공정명", values="공정시간", aggfunc="mean").round(1)

#피벗테이블에서 값 가져오기
machin = pivot.loc["설비"]
stack_machin = machin.stack()
max_time_machin = machin.max().max()
max_machin_data = pivot.loc["설비"][pivot.loc["설비"] == max_time_machin].stack().reset_index()
max_machin_data.columns = ["시간", "공정명", "공정시간"]
display(max_machin_data)


human = pivot.loc["작업자"]
stack_human = human.stack()
max_time_human = human.max().max()
max_human_data = pivot.loc["작업자"][pivot.loc["작업자"] == max_time_human].stack().reset_index()
max_human_data.columns = ["시간", "공정명", "공정시간"]
display(max_human_data)


```
