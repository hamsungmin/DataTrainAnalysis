### 목표1
Python으로 week1_project_data.xlsx 파일을 열어서 '가동실적' 시트에 접근하세요

```python
from openpyxl import load_workbook

wb = load_workbook('week1_project_data.xlsx')
ws = wb['가동실적']
```

### 목표2
전체 공정의 ‘공정시간’의 평균을 구하세요

```python
row = 3

target_id = 20006

tat_lst = []
while True:
  tat = ws[f"F{row}"].value
  if tat == None:
    break
  tat_lst.append(tat)
  row += 1

print("전체 행 수:", row)
```

```python
mean_tat = sum(tat_lst) / len(tat_lst)
step2_answer = round(mean_tat, 2)
print(f"평균 공정시간: {step2_answer}")
```

### 목표3
'공정 ID'가 20006인 공정의 '공정시간'의 평균을 구하세요

```python
row = 3

target_id = 20006

tat_lst = []
while True:
  _id = ws[f"D{row}"].value
  if _id == target_id:
    tat = ws[f"F{row}"].value
    tat_lst.append(tat)
  elif _id == None:
    break
  row += 1

print("전체 행 수:", row)
```

```python
mean_tat = sum(tat_lst) / len(tat_lst)
step3_answer = round(mean_tat, 2)
print(f"공정시간: {step3_answer}")
```

### 목표4
'장비이름'이 'OP40조립기' 이고 공정 종류가 '작업자'인 공정 중 '공정시간'이 최대인 공정을 구하고, 해당 공정의 '공정시간', '시작시간', '종료시간'을 구하세요
</br>
(최대인 공정이 여러 개 있다면 해당하는 모든 공정의 시작시간과 종료시간을 구하세요)

```python
row = 3

target_equipment = "OP40조립기"
target_type = "작업자"

max_tat = 0
step4_answer_lst = []

while True:
  equipment = ws[f"A{row}"].value
  if equipment == target_equipment:
    _type = ws[f"I{row}"].value
    if _type == target_type:
      tat = ws[f"F{row}"].value

      if tat >= max_tat:
        if tat > max_tat:
          step4_answer_lst.clear()
          max_tat = tat
        max_tat_info = {}
        max_tat_info["tat"] = tat
        max_tat_info["start_time"] = ws[f"G{row}"].value
        max_tat_info["end_time"] = ws[f"H{row}"].value
        step4_answer_lst.append(max_tat_info)

      tat_lst.append(tat)
  elif equipment == None:
    break
  row += 1

print("전체 행 수:", row)
```

```python
for max_tat_info in step4_answer_lst:
  print(f"공정시간: {max_tat_info['tat']}, 시작시간: {max_tat_info['start_time']}, 종료시간: {max_tat_info['end_time']}")
```

### 목표5
'공정명'이 '조립검사'이고 '시작시간'과 '종료시간'이 모두 2022-08-02 08:00:00 ~ 2022-08-02 09:00:00 사이인 공정의 '공정시간'의 평균을 구하세요
</br>
※ 권장 라이브러리: datetime
</br>
※ 구글링으로 datetime 라이브러리의 사용법을 찾아보시는 것을 권장드립니다

```python
from datetime import datetime

row = 3
target_name = "조립검사"
min_time = datetime.strptime("2022-08-02 08:00:00", "%Y-%m-%d %H:%M:%S")
max_time = datetime.strptime("2022-08-02 09:00:00", "%Y-%m-%d %H:%M:%S")

tat_lst = []

while True:
  name = ws[f"E{row}"].value
  if name == target_name:
    start_time = ws[f"G{row}"].value
    end_time = ws[f"H{row}"].value
    try:
      start_time = datetime.strptime(start_time, "%Y-%m-%d %H:%M:%S.%f")
    except ValueError:
      start_time = datetime.strptime(start_time, "%Y-%m-%d %H:%M:%S")
    try:
      end_time = datetime.strptime(end_time, "%Y-%m-%d %H:%M:%S.%f")
    except ValueError:
      end_time = datetime.strptime(end_time, "%Y-%m-%d %H:%M:%S")

    if (min_time <= start_time <= max_time) and (min_time <= end_time <= max_time):
      tat = ws[f"f{row}"].value
      tat_lst.append(tat)

  elif name is None:
    break

  row += 1

print("전체 행 수:", row)
```

```python
mean_tat = sum(tat_lst) / len(tat_lst)
step5_answer = round(mean_tat, 2)
print(f"평균 공정시간: {step5_answer}")
```

### 목표7
목표6에서 저장한 텍스트 파일을 Python을 이용하여 이메일에 첨부하고 이메일 전송해주세요

※ 권장 라이브러리: smtplib, email

```python
from email.mime.multipart import MIMEMultipart

message = MIMEMultipart()
# 이메일 발신자
sender_email = "abc@gmail.com"
message["From"] = sender_email

# 이메일 수신자
recipient_email = "sss@gmail.com"
message["To"] = recipient_email
```

```python
from email.mime.text import MIMEText

# 이메일 제목
message["Subject"] = "1주차 프로젝트"

# 이메일 내용
body = "안녕하세요.\n1주차 프로젝트 파일 첨부합니다."
message.attach(MIMEText(body, "plain"))
```

```python
from email.mime.base import MIMEBase
from email import encoders

with open(report_path, "rb") as attachment:
  part = MIMEBase("application", "octet-stream")
  part.set_payload(attachment.read())

encoders.encode_base64(part)

part.add_header(
  "Content-Disposition",
  "attachment",
  filename=report_path
)

message.attach(part)
```

```python
import smtplib

# SMTP 서버 설정
smtp_server = "smtp.gmail.com"
smtp_port = 587

################## 각자 발급 필요 ##################
password = "your_password"
####################################################

with smtplib.SMTP(smtp_server, smtp_port) as server:
  server.starttls()
  server.login(sender_email, password)
  # 이메일 전송
  server.sendmail(sender_email, recipient_email, message.as_string())
```
